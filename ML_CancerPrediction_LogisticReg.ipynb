**CANCER PREDICTION**

For this cancer prediction task, the goal of the model is likely to predict whether a given set of predictors indicate that a tumor is Malignant (M) or Benign (B).

```python
import pandas as pd
```

```python
cancer = pd.read_csv('https://github.com/YBIFoundation/Dataset/raw/main/Cancer.csv')
```

```python
cancer.head(20)
```

```python
cancer.info()
```

```python
cancer.describe()
```

```python
cancer.columns
```

```python
y =cancer['diagnosis']
```

```python
X=cancer.drop(['id','diagnosis','Unnamed: 32'],axis=1)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, train_size=0.7, random_state=2529)
```

```python
# check shape of train and test sample
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression(max_iter=5000)
```

```python
model.fit(X_train,y_train)
```

```python
model.intercept_
```

```python
model.coef_
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
confusion_matrix(y_test,y_pred)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

```python
# Assuming y_pred is an array of predictions
y_pred_df = pd.DataFrame(y_pred, columns=['Predicted'])

# Now you can count the occurrences of each unique value
counts = y_pred_df['Predicted'].value_counts()

# Print the counts
print(counts)
```
